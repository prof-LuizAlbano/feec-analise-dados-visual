# **Teste de Hipóteses - Qui-Quadrada**

```python
# Adicionar a coluna de evasão binária
df_filtered['Evasao'] = df_filtered['Situação de Matrícula'].apply(lambda x: "Evadido" if x in ['Abandono', 'Cancelada', 'Desligada'] else "Não Evadido")
```

```python
chi = pd.crosstab(df_filtered["Renda Familiar"], df_filtered["Evasao"])
chi
```

```python
# Cálculo do Qui-Quadrado a partir da tabela de contingência
chi_scores = chi2_contingency(chi)
chi_scores
```

```python
def qui_quadrado(df, variavel, lista):

  # Calcular Qui2 e p-valor para cada variável e colocar resultado num df
  d = {'Qui2': [], 'p-Valor': []}

  for l in lista:

      # Cria tabela que relaciona variável com lista
      tab = pd.crosstab(df[variavel], df[l])

      # Calcula o qui2 entre as duas variáveis
      chi_scores = chi2_contingency(tab)

      # Recupera o valor e p-value do teste
      scores = pd.Series(chi_scores[0])
      pvalues = pd.Series(chi_scores[1])

      # Adiciona resultado ao dicionário
      d['Qui2'].append(scores[0])
      d['p-Valor'].append(pvalues[0])

  # Transformar dicionário em dataframe
  chi_squared = pd.DataFrame(d)

  # Formato de visualização dos números
  pd.options.display.float_format = "{:,.2f}".format

  # Renomear linhas do df
  for n, l in zip(np.arange(0, len(lista), 1), lista[0:len(lista)]):
    chi_squared = chi_squared.rename(index = {n: l})

  # Mostrar em ordem crescente para o p-valor
  chi_squared = chi_squared.sort_values(by = 'p-Valor', ascending = True)

  # Visualizar df
  return chi_squared
```

```python
vars = ['Renda Familiar', 'Cor/Raça', 'Faixa Etária', 'Sexo', 'Tipo de Curso']

qui_quadrado(df_filtered, "Evasao", vars)
```

```python
def analise_bi(df, var1, var2, ordem_linhas, ordem_colunas):

    df_group = df.groupby([var1, var2])[var2].count().unstack(var2).fillna(0)

    df_group = df_group.reindex(ordem_linhas)

    # Reordenar colunas
    df_group = df_group[ordem_colunas]

    return df_group
```

# **Teste de Hipóteses**

```python
# 1. Converter os intervalos de RFP dado em intervalo de valores como valores
# númericos Reais, Atribuindo para cada caso a média
# respectiva do seu intervalo e Acrescentar uma partição desse dataset
# "df_evaded" em duas categorias principais, sobre as quais se incidirão a análise
# Renda Familiar Abaixo da Média: '0<RFP<=0,5' +  '0,5<RFP<=1' + '1<RFP<=1,5'
# Renda Familiar Acima da Média:  '1,5<RFP<=2,5' +  '2,5<RFP<=3,5' +  'RFP>3,5'

#Salário Mínimo Nominal de 2020, segundo o link https://www.dieese.org.br/analisecestabasica/salarioMinimo.html#2020
salario_min_nominal = 1.045

# 1.a) Mapeamento dos intervalos para médias
intervalos = {
    '0<RFP<=0,5': 0.25,
    '0,5<RFP<=1': 0.75,
    '1<RFP<=1,5': 1.25,
    '1,5<RFP<=2,5': 2.0,
    '2,5<RFP<=3,5': 3.0,
    'RFP>3,5': 4.0,
    'Não declarada': None  # Se não declarada, pode ser tratada como NaN ou removida
}

# 1.b) Substituir os intervalos pelas médias calculadas
df_filtered['Renda_Familiar_Media'] = df_filtered['Renda Familiar'].map(intervalos)

# 1.c) Remover linhas com NaN na coluna de renda familiar média (caso aplicável)
df_filtered.dropna(subset=['Renda_Familiar_Media'], inplace=True)

# 1.d) Acrescentar uma coluna de categoria da renda familiar, dividindo os dados em Estudantes de Renda Familiar Abaixo da Média e Acima da Média, considerando 1.6 como a média da Renda Familiar
df_filtered['Categoria_Renda_Familiar_Media'] = df_filtered['Renda_Familiar_Media'].apply(lambda x: 'baixa_renda' if x < salario_min_nominal else 'alta_renda')

grupo_baixa_renda = df_filtered[df_filtered['Categoria_Renda_Familiar_Media'] == 'baixa_renda']
grupo_alta_renda = df_filtered[df_filtered['Categoria_Renda_Familiar_Media'] == 'alta_renda']
```

```python
# 2. Calcular a taxa de evasão para cada grupo
taxa_evasao_baixa_renda = (grupo_baixa_renda['Evasao'] == 'Evadido').mean()
taxa_evasao_alta_renda = (grupo_alta_renda['Evasao'] == 'Evadido').mean()
observed_diff = taxa_evasao_baixa_renda - taxa_evasao_alta_renda

# 3. Gerar amostras bootstrap para cada grupo
n_iterations = 10000
diffs = []

for _ in range(n_iterations):
    sample_baixa_renda = resample(grupo_baixa_renda)
    sample_alta_renda = resample(grupo_alta_renda)
    sample_diff = (sample_baixa_renda['Evasao'] == 'Evadido').mean() - (sample_alta_renda['Evasao'] == 'Evadido').mean()
    diffs.append(sample_diff)
```

```python
# 4. Obter a distribuição das diferenças e calcular o intervalo de confiança
diffs = np.array(diffs)
ci_lower = np.percentile(diffs, 2.5)
ci_upper = np.percentile(diffs, 97.5)

# 5. Calcular o valor-p
p_value = np.mean(diffs >= observed_diff)

# 6. Verificar se o intervalo de confiança inclui zero e interpretar o valor-p
print(f"Observed difference: {observed_diff}")
print(f"95% Confidence Interval: [{ci_lower}, {ci_upper}]")
print(f"p-value: {p_value}")

if ci_lower <= 0 <= ci_upper:
    print("Não rejeitamos a hipótese nula: a renda familiar não influencia a evasão dos alunos.")
else:
    print("Rejeitamos a hipótese nula: a renda familiar influencia a evasão dos alunos.")

if p_value < 0.05:
    print("O valor-p é significativo ao nível de 5%. Rejeitamos a hipótese nula.")
else:
    print("O valor-p não é significativo ao nível de 5%. Não rejeitamos a hipótese nula.")

# 7. Plotar o resultado
plt.hist(diffs, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(observed_diff, color='red', linestyle='dashed', linewidth=2, label='Diferença Observada')
plt.axvline(ci_lower, color='blue', linestyle='dashed', linewidth=2, label='Limite Inferior IC 95%')
plt.axvline(ci_upper, color='blue', linestyle='dashed', linewidth=2, label='Limite Superior IC 95%')
```

### **1. Estruturação da análise para responder a questão inicial do projeto**

Primeiro, precisamos formular nossas hipóteses nula e alternativa:

**Hipótese Nula (H₀):** Não há diferença significativa nas proporções de evasão entre estudantes de diferentes categorias de renda familiar.

**Hipótese Alternativa (H₁):** Há uma diferença significativa nas proporções de evasão entre estudantes de diferentes categorias de renda familiar.

**Estatística de Interesse:** A diferença nas proporções de evasão entre duas categorias de renda familiar, por exemplo, "abaixo_1_6" e "acima_1_6".

**Objetivo do Estudo:** Queríamos saber se há uma diferença significativa nas proporções de evasão entre estudantes de diferentes categorias de renda familiar.

**Método Utilizado:** Através de uma função usamos a técnica bootstrap para reamostrar nossos dados até 10.000 vezes e calcular a diferença nas proporções de evasão nas diferentes categorias de Renda Familiar em cada uma das amostras.

### **2. Resultados Obtidos**

**p-Valor:** 0.4862

**Intervalo de Confiança (95%):** [0.45, 0.91]

**Diferença Observada:** 0.72

### **3. Interpretação dos resultados**

**Intervalo de Confiança (IC):**

**Inclui zero:** Não há diferença significativa (não rejeitamos a hipótese nula).<br>
**Não inclui zero:** Há diferença significativa (rejeitamos a hipótese nula).<br>
**IC(95%):** [0.0007983015358053152, 0.013713710564814699]

**Valor-p:**

**Menor que 0.05:** Diferença é significativa (rejeitamos a hipótese nula).<br>
**Maior ou igual a 0.05:** Diferença não é significativa (não rejeitamos a hipótese nula).<br>
No nosso caso o **valor-p** é: 0.494

Ambos os testes (**IC e valor-p**) fornecem uma forma de verificar a significância estatística da diferença observada nas taxas de evasão entre os grupos de renda alta e baixa. Combinar essas duas análises é muito interessante, nos dando maior confiança nas decisões sobre as hipóteses. Repara que pelo intervalo de confiança podemos ver que a Renda Familiar influencia o nível de evasão de estudantes, rejeitando a Hipótese nula (H0), mas pelo valor-p percebemos que não há significância estátistica suficiente para rejeitar a H0, o que sugere que talvez a Renda Familiar apesar de poder influenciar no nível de envasão, mas não é tão determinante, o que faz necessário analisar outros factores que podem combinados mensurar melhor as chances de envasão dos estudantes

Partindo de um bechmark de 0.05 para o p-valor, as variáveis **Faixa Etária**, **Tipo de Curso**, **Sexo** e **Renda Familiar**, estão relacionados com a Evasão dos estudantes.
