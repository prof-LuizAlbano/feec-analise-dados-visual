```python
from plotnine import *
import pandas as pd
import numpy as np
from scipy.stats import norm

# Definindo os limites do eixo x
xlim = (-6, 6)

# Criando um DataFrame com valores de x dentro dos limites especificados
x_values = np.linspace(xlim[0], xlim[1], 101)
df = pd.DataFrame({'x': x_values})
```

```python
p1 = (
    ggplot(df, aes(x='x'))
    + stat_function(fun=norm.pdf, args={'loc': 0, 'scale': 0.5})
    + ylab('') 
    + geom_vline(xintercept=0, linetype='dotted')
    + coord_cartesian(xlim=xlim, ylim=(0, 0.8))
    + ggtitle('High precision, high accuracy')
)

p1
```

```python
p2 = (
    ggplot(df, aes(x='x'))
    + stat_function(fun=norm.pdf, args={'loc': 2.5, 'scale': 0.5})
    + ylab('')
    + geom_vline(xintercept=0, linetype='dotted')
    + coord_cartesian(xlim=xlim, ylim=(0, 0.8))
    + ggtitle('High precision, low accuracy')
)

p2
```

```python
p3 = (
    ggplot(df, aes(x='x'))
    + stat_function(fun=norm.pdf, args={'loc': 0, 'scale': 1.5})
    + ylab('')
    + geom_vline(xintercept=0, linetype='dotted')
    + coord_cartesian(xlim=xlim, ylim=(0, 0.8))
    + ggtitle('Low precision, high accuracy')
)

p3
```

```python
p4 = (
    ggplot(df, aes(x='x'))
    + stat_function(fun=norm.pdf, args={'loc': 2.5, 'scale': 1.5})
    + ylab('')
    + geom_vline(xintercept=0, linetype='dotted')
    + coord_cartesian(xlim=xlim, ylim=(0, 0.8))
    + ggtitle('Low precision, low accuracy')
)

p4
```

```python
# Combine os gráficos usando facet_wrap para colocá-los lado a lado
combined_plot = (p1 + p2 + p3 + p4) + facet_wrap("~ggtitle", nrow=2)

# Exibe o gráfico combinado
print(combined_plot)
```
