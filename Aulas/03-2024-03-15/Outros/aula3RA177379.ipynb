```python
algebra = diamonds[['carat', 'cut', 'price']]
algebra
```

```python
ggplot(data = algebra) + geom_point(mapping = aes(x=algebra['price'],y=algebra['carat'], color = algebra['cut']))
```

```python
ggplot (data = algebra) + geom_point (mapping = aes(x=algebra['price'],y=algebra['carat'], color = algebra['cut'])) + scale_x_continuous (breaks = range(2500,17500,2500)) + scale_y_continuous (breaks = [x / 10.0 for x in range(5, 50, 15)])
```

```python
print('u_carat =', algebra['carat'].mean())
print('sd_carat =', algebra['carat'].std())
print('Md_carat =', algebra['carat'].median())

print('u_price = U$', algebra['price'].mean())
print('sd_price = U$', algebra['price'].std())
print('Md_price = u$', algebra['price'].median())
```

```python
ggplot (data = algebra) + geom_histogram (mapping = aes(x=algebra['price'], color = algebra['cut']), binwidth=2000)
```

```python
ggplot (data = algebra) + geom_freqpoly (mapping = aes(x=algebra['price'], color = algebra['cut']), binwidth=2000)
```

```python
ggplot (data = algebra) + geom_histogram (mapping = aes(x=algebra['price'], color = algebra['cut']), binwidth=2000) + coord_flip ()
```

```python
ggplot (data = algebra) + geom_histogram (mapping = aes(x=algebra['price'], color = algebra['cut']), binwidth=2000)
# + coord_polar()
```

```python
ggplot (data = algebra, mapping = aes(x=algebra['price'],y=algebra['carat'], color = algebra['cut'])) + geom_point (mapping = aes()) + geom_smooth(method = "lm", se = False, color = 'blue')
```

```python
ggplot (data = algebra, mapping = aes(x=algebra['price'],y=algebra['carat'], color = algebra['cut'])) + geom_point () + facet_wrap ('~cut', nrow=2)
```
